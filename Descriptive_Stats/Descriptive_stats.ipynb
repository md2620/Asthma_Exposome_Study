MORE CLARITY WITH BOXPLOTS ADDED

```python
continous_var = ['RIDAGEYR']
plt.figure(figsize=(16, 25))
for i, col in enumerate(continous_var):
    plt.subplot(6, 4, i*2+1)
    plt.subplots_adjust(hspace =.25, wspace=.3)
    
    plt.grid(True)
    plt.title(col)
    sns.kdeplot(df_subset.loc[df_subset["casecont"]=='0', col], label="No Asthma attack last year", color = "firebrick", shade=True, kernel='gau', cut=0)
    sns.kdeplot(df_subset.loc[df_subset["casecont"]=='1', col], label="Asthma attack last year",  color = "teal", shade=True, kernel='gau', cut=0)
    plt.subplot(6, 4, i*2+2) 
    sns.boxplot(y = col, data = df_subset, x="casecont", palette = ["firebrick", "teal"])
plt.savefig('/rds/general/user/md2620/home/ML_project/Plots/desc_stat_aft_2')
```
